```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/Users/omgajipara/Desktop/Data Science/AI:ML/Machine Learning/Unsupervised Learning/Clustering/K-Means Clustering/iris.csv")
data
```

```python
New_Data=data.drop(['species'],axis=1)
New_Data
```

```python
sns.pairplot(data=New_Data)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(New_Data, method='single', metric='euclidean'))
plt.title("Dendrogram - Single Linkage")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")

plt.show()
```

```python

plt.figure(figsize=(10, 10))  # Place this BEFORE plotting

# Create dendrogram
dendrogram = sc.dendrogram(sc.linkage(New_Data, method='ward'))

# Set a visible threshold line
plt.axhline(y=60, color='r', linestyle='--') 
print(plt.ylim())  # This prints the y-axis limits to console

# Add labels and title
plt.title("Dendrogram - Ward Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")

# Save the figure

# Show the figure
plt.show()
```

```python
plt.figure(figsize=(8, 5))  # Place this BEFORE plotting

# Create dendrogram
dendrogram = sc.dendrogram(sc.linkage(New_Data, method='ward'))

# Set a visible threshold line
plt.axhline(y=60, color='r', linestyle='--') 
print(plt.ylim())  # This prints the y-axis limits to console

# Add labels and title
plt.title("Dendrogram - Ward Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")

# Save the figure
plt.savefig('assets/HierarchicalClustering.jpg')

# Show the figure
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
New_Data['Cluster'] = model.fit_predict(New_Data)
```

```python
print(New_Data.columns)
```

```python


sns.pairplot(data=New_Data,hue='Cluster')
plt.title("Hierarchical Clustering")
plt.savefig('assets/HierarchicalClustering1.jpg')
plt.show()
```
